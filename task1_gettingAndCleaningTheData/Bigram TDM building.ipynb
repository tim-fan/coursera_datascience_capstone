```r
#Use a 10k tweet subset for testing
enUsTwitter10kSubset <- readRDS(file='../data/subsets/enUsTwitter10kSubset.rds')
head(enUsTwitter10kSubset)
```

```r
# Following https://gist.github.com/benmarwick/5370329
library("RWeka")
library("tm")

df <- data.frame(x=unlist(matrix(enUsTwitter10kSubset)), stringsAsFactors = F)

corpus <- Corpus(VectorSource(df$x)) 
# corpus <- tm_map(corpus, PlainTextDocument)

BigramTokenizer <- function(x) NGramTokenizer(x, Weka_control(min = 2, max = 2))
tdm <- TermDocumentMatrix(corpus, control = list(tokenize = BigramTokenizer))

inspect(tdm)
```
